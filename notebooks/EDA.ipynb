# Fake Job Offer Detection - Exploratory Data Analysis (EDA)
### Domain: Student Career Safety (Applied Artificial Intelligence)

This notebook performs robust exploratory data analysis on the job offer scams dataset created based on **FTC Job Scam Guidance** red flags. We will examine indicators like upfront payment requests, unrealistic salaries, extreme urgency, and suspicious domains, analyzing class distributions, text lengths, and word frequencies.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re

# Set style
sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

## 1. Load Synthetic Labeled Dataset based on FTC Guidance
We extract indicators from the FTC job scam portal to structure a baseline training dataset.

```python
data = [
    {
        "offer_text": "URGENT hiring! Earn $150 per hour doing simple data entry from home. No experience needed. Start immediately. You must pay a $45 registration fee for our certified software package. Send money via wire transfer, Western Union, or Bitcoin to reserve your slot now. Interview conducted on Telegram app.",
        "company_domain": "telegram-jobs-now.com",
        "asks_payment": 1,
        "salary_claim": "suspicious",
        "urgency": "high",
        "risk_label": 1
    },
    {
        "offer_text": "Dear candidate, you have been instantly selected for the Virtual Assistant position. No interview is required. The salary is $5,000 weekly, paid via check. Before receiving your laptop, you need to pay $100 for a security background check which will be fully refunded in your first salary. Please buy steam or apple gift cards for verification.",
        "company_domain": "gmail.com",
        "asks_payment": 1,
        "salary_claim": "suspicious",
        "urgency": "high",
        "risk_label": 1
    },
    {
        "offer_text": "We are looking for a Software Engineer Intern to join our engineering team. The ideal candidate has experience with React, Node.js, and SQL. This is a paid summer internship offering $25/hour. The selection process involves a technical assessment followed by a panel interview. No fees or deposits required.",
        "company_domain": "stripe.com",
        "asks_payment": 0,
        "salary_claim": "realistic",
        "urgency": "normal",
        "risk_label": 0
    },
    {
        "offer_text": "Google is recruiting a summer UX Design Intern. Work closely with product managers and engineers to craft user flows, wireframes, and high-fidelity mockups. Competitive stipend provided. Requires active enrollment in a Bachelor's or Master's design program. Apply online on our corporate jobs site.",
        "company_domain": "google.com",
        "asks_payment": 0,
        "salary_claim": "realistic",
        "urgency": "normal",
        "risk_label": 0
    }
]

df = pd.DataFrame(data)
print(f"Dataset Shape: {df.shape}")
df.head()
```

## 2. Text Analysis: Word Count & Character Count
Let's compute simple characteristics of our text features.

```python
df["char_count"] = df["offer_text"].apply(len)
df["word_count"] = df["offer_text"].apply(lambda x: len(x.split()))
df.groupby("risk_label")[["char_count", "word_count"]].mean()
```

## 3. Visualize Feature Distributions
We look at how our class labels relate to specific red flags.

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Risk Label Distribution
sns.countplot(data=df, x="risk_label", ax=axes[0], palette="Set2")
axes[0].set_title("Distribution of Genuine (0) vs. Fraudulent (1) Job Offers")
axes[0].set_xticklabels(["Genuine", "Scam/Fraud"])

# Upfront Payment Requests by Class
sns.barplot(data=df, x="risk_label", y="asks_payment", ax=axes[1], palette="coolwarm")
axes[1].set_title("Proportion of Offers Requesting Upfront Payment")
axes[1].set_xticklabels(["Genuine", "Scam/Fraud"])
plt.tight_layout()
plt.show()
```
