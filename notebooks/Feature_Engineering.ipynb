# Fake Job Offer Detection - NLP & Feature Engineering
### Domain: Student Career Safety (Applied Artificial Intelligence)

This notebook covers the full text processing pipeline used to convert raw unstructured job offer text into standardized numerical feature vectors for ML model feeding.

```python
import re
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

documents = [
    "URGENT hiring! Earn $150 per hour doing simple data entry. Pay registration fee.",
    "Dear applicant, you are selected. Please pay security deposit.",
    "We are looking for a Software Engineer Intern. React, Node.js, SQL experience.",
    "Google is recruiting UX Design Interns. Competitive stipend. Apply online."
]
```

## 1. Custom Text Preprocessing Helper
Let's write a reusable, robust cleaning utility that normalizes terms.

```python
def clean_and_tokenize(text):
    # Lowercase & strip punctuation
    cleaned = re.sub(r'[^\w\s]', ' ', text.lower())
    # Split on whitespace
    tokens = cleaned.split()
    # Simple stopword removal
    stopwords = {"i", "me", "my", "we", "our", "you", "your", "he", "she", "it", "they", "the", "and", "or", "to", "for", "in", "of", "on", "at", "is", "are", "am", "was", "were", "be", "been", "being"}
    filtered_tokens = [t for t in tokens if len(t) > 2 and t not in stopwords]
    return " ".join(filtered_tokens)

processed_docs = [clean_and_tokenize(doc) for doc in documents]
for orig, proc in zip(documents, processed_docs):
    print(f"Original: {orig}\nProcessed: {proc}\n")
```

## 2. TF-IDF Vectorization
Using Scikit-Learn's TfidfVectorizer to create numerical arrays representing term frequencies weighted by their corpus importance.

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(processed_docs)

tfidf_df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
print("TF-IDF Feature Names:", list(tfidf_df.columns))
tfidf_df
```
